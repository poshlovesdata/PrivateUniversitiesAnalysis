## Exploratory Data Analysis

```python
uni_df2 = pd.read_csv("private_uni_cleaned.csv")
```

```python

uni_df2
```

### What are the top 5 universities by year established?

```python
top_uni = uni_df2.sort_values(by="Year Established").head(5)
# top_uni
top_uni[["University Name","State","Year Established"]]
```

### How many universities are there in each state?

```python
top_states = uni_df2.groupby("State")["University Name"].count()
sorted_top_states = top_states.sort_values(ascending=False)
sorted_top_states
```

### What is the average fee for universities in each region?


```python
avg_region = uni_df.groupby("Region")["Fees"].mean()
avg_region_standard = avg_region.apply(lambda x: "{:.2f}".format(x))
avg_region_standard
```

### Which state has the highest number of universities?

```python
sorted_top_states.head(1)
```

### How many universities were established in each year?

```python
year_uni_count = uni_df2.groupby("Year Established")["University Name"].count()
year_uni_count.sort_values(ascending=False)
```

### What is the total fee collected by universities in each region?

```python
uni_df2.groupby("Region")["Fees"].sum().sort_values(ascending=False)
```

### Which university has the highest fee?

```python
uni_df2.sort_values("Fees", ascending=False).head(1)
```

### How many universities are there in each region?


```python
uni_df2.groupby("Region")["University Name"].count().sort_values(ascending=False)
```

### What is the average fee for universities in each state?

```python
avg_state_fee = uni_df2.groupby("State")["Fees"].mean().sort_values(ascending=False)

avg_state_fee.apply(lambda x: "{:.2f}".format(x))
```

```python
sorted_top_states["Bayelsa"]
```

### What is the average fee for universities in each state witht the count?

```python
state_group = uni_df2.groupby("State").agg(
    avg_fee = ("Fees", "mean"),
    uni_count = ("University Name", "count")
)

state_group['avg_fee'] = state_group["avg_fee"].apply(lambda x: "{:.2f}".format(x))
# state_group
state_group.sort_values(by="avg_fee", ascending=False)
```
